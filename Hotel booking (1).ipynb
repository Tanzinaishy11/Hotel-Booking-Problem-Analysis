# Data Anlysis and Visualization

```python
df.head()
```

```python
#find the percantage of cancellation

cancelled_perc = df['is_canceled'].value_counts(normalize = True)
cancelled_perc
```

```python
#63% booking not cancelled & 37% booking cancelled
```

```python
#data Visualization using bar chart
print(cancelled_perc)
plt.figure(figsize = (5,5))
plt.title('Reservation Status count')
plt.bar(['not cancelled','cancelled'],df['is_canceled'].value_counts())
plt.show()
```

```python
#data visualization using pie chart
plt.figure(figsize = (5,5))
plt.pie(df['is_canceled'].value_counts(), labels=['not cancelled', 'cancelled'], autopct='%1.1f%%', startangle=90)
plt.title('Reservation Status count')
plt.axis('equal') 
plt.show()
```

```python
#analys the reservation status of each hotel

plt.figure(figsize=(10, 6))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status in Different Hotels', size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of Reservations')
plt.legend(['Not canceled','Cancelled'])
plt.show()
```

#according to the ratio resort hotels cancellation rate higher than city hotel.

```python
resort_hotel = df[df['hotel']=='Resort Hotel'] #filtering rows from original dataframe
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
City_hotel = df[df['hotel']=='City Hotel']
City_hotel['is_canceled'].value_counts(normalize = True)
```

```python
#impact of price on cancellation rate
#summarizing data to find the mean average daily rate for each reservation status date in the 'resort_hotel' & 'city_hotel' category
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
City_hotel = City_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label = 'Resort Hotel')
plt.plot(City_hotel.index, City_hotel['adr'],label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

So the price of city hotel is less than resort hotel for some of the days

```python
#cancellation rate according to month
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='Blues')
legend_labels, _ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status per month', size=20)
plt.xlabel('month')
plt.ylabel('Number of Reservations')
plt.legend(['Not canceled','Cancelled'])
plt.show()
```

In january cancelation rate is highest.
In August & september cancellation rate is lowest where reservation rate is highest August.


```python
#checking does price has any impact on cancellation rate
#plotting ADR for each month

# Filter the DataFrame for canceled reservations and group by month to calculate the sum of ADR per month
canceled_reservations = df[df['is_canceled'] == 1] #data filtering
monthly_adr = canceled_reservations.groupby('month')['adr'].sum().reset_index() #data aggregation

# Create a bar plot to show ADR per month
plt.figure(figsize=(15, 8)) #data visualization
plt.title('ADR per month', fontsize=30)
sns.barplot(x='month', y='adr', data=monthly_adr)
plt.show()

```

According to the 'reservation rate per month' and 'ADR per month' graph-
1. September - ADR lowest, Cancellation rate lowest
2. Januray - ADR highest, Cancellation rate highest
so hotel room price has the direct impact on cancellation rate.


```python
#cancelation rate based on countries (for top 10)
cancelled_data = df[df['is_canceled']==1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 coutries with reservation canceled')
plt.pie(top_10_country, autopct = '%.1f',labels = top_10_country.index)
plt.show()
```

portugal has the highest cancellation rate. so all the hotels PRT needs to take care of their systems and facilities and lower the rate of booking. They may advertise and run campaigns about their hotel to decrease the cancellatiion rate.
